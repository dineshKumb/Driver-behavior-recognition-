This item concerns driver behaviour recognition (specifically, the normal, labelled as “1”, and abnormal driving behaviours, labelled as “2”) using data collected in 2016 by using a smartphone app DriveSafe for the purpose of deep driving behaviour analysis. The app can capture vehicle-related measurements such as speed, acceleration, gyroscope, and driver status (normal or abnormal).

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

importing all the required libraries

```python
dataset_X_train = pd.read_csv('X_train.csv')
X_train= dataset_X_train.iloc[:, :].values
dataset_y_train = pd.read_csv('y_train.csv')
y_train= dataset_y_train.iloc[:, :].values
X_test = pd.read_csv('X_test.csv')
y_test= pd.read_csv('y_test.csv')
```

here i have imported the DecisionTreeRegressor class from the sklean library then by giving it an instance i have used this class to fit the traing data files.

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)
```

To check how well our module has trained, i have used the predict class from the regressor to predict the values for the given X_test file.

```python
y_prediction= regressor.predict(X_test)
```

```python
y_prediction
```


    To predict the accuracy of the trained module, here i made a comparison between the actual values given in y_test and the predicted values saved in the y_prediction  

```python
#How often is the classifier correct, according to the model?
print("Accuracy:",metrics.accuracy_score(y_test, y_prediction))
```

the accuracy of this ddecision tree regression module is 76.3237%
